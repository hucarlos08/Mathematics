# Plot in python

**Step-by-Step Guide to Graphing a Function**

1. **Define the Mathematical Function**: Clearly state the mathematical function you intend to graph. Ensure that the function is well-defined over the domain you wish to explore. For instance, $f(x) = x^2$ is a simple quadratic function.

2. **Determine the Domain**: Decide the range of input values (domain) for which you want to visualize the function. Consider any limitations of the function, such as asymptotes or discontinuities.

3. **Choose an Appropriate Resolution**: Decide on the resolution or the number of points you want to calculate between the minimum and maximum values of your domain. Higher resolution will result in smoother curves but may take longer to compute.

4. **Generate Input Values**: Use a tool like NumPy to create an array of input values (`x`) that are evenly spaced across your chosen domain. For instance, `np.linspace(start, end, num_points)` can be used to generate these values.

5. **Compute Output Values**: Apply the mathematical function to each input value to get the corresponding output values (`y`). This step may involve vectorized computation if using NumPy for efficiency.

6. **Set Up the Plotting Environment**: Import a plotting library such as Matplotlib and prepare to create a plot by initializing figure and axis objects.

7. **Plot the Function**: Utilize the plotting commands such as `plt.plot(x, y)` to create a 2D line plot of the function. Adjust line styles, colors, and markers as needed for clarity.

8. **Customize the Graph**: Add titles, axis labels, and a legend if needed. Customize the axis limits, ticks, and grid lines to enhance readability.

9. **Analyze Continuity and Differentiability (if needed)**: If your function has complex behavior, consider plotting derivative or integral functions alongside the original to provide more insight.

10. **Create Multiple Plots for Comparison (if needed)**: If comparing functions, you may want to create subplots or multiple figures to visualize differences or similarities clearly.

11. **Save or Display the Plot**: Use `plt.show()` to display the plot interactively or `plt.save('filename.ext')` to save the figure to a file.

12. **Interpret the Results**: Take a moment to analyze the graph. Look for patterns, trends, or anomalies that arise from the visualization. Note any behavior such as convergence, divergence, or periodicity.

Remember that this process may involve iteration. You might need to go back and adjust your domain, resolution, or visualization parameters based on initial results to get the most informative graph possible.

# Lets start with the simple 1D-plot

```python
# Importing the necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Define the function to plot. In this case, we are using a simple linear function.
def f(x):
    return 2 * x + 1

# Generate a range of values from -10 to 10 to represent the x-axis
x = np.linspace(-10, 10, 400)

# Apply the function to each element in the x array to get the y values
y = f(x)

# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(x, y, '-r', label='$y=2x+1$')

# Adding title and labels
plt.title('Graph of $y=2x+1$')
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
# Importing necessary libraries for 3D plotting
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Define the function that we want to plot (e.g., z = x^2 + y^2)
def f(x, y):
    return x**2 + y**2

# Create a grid of values (x, y) using NumPy's meshgrid function
N =  100
x = np.linspace(-5, 5, N)
y = np.linspace(-5, 5, N)
x, y = np.meshgrid(x, y)

# Apply the function to each point in the grid
z = f(x, y)

# Set up the figure and 3D axis
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Add a color bar which maps values to colors
fig.colorbar(surf, shrink=0.5, aspect=5)

# Adding labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Surface Plot of z=x^2+y^2')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function (as before)
def f(x, y):
    return x**2 + y**2

# Define the gradient of the function
def grad_f(x, y):
    return 2*x, 2*y

# Create a grid of values (x, y)
N = 20
x = np.linspace(-5, 5, N)
y = np.linspace(-5, 5, N)
x, y = np.meshgrid(x, y)

# Calculate function values
z = f(x, y)

# Calculate gradient values
gx, gy = grad_f(x, y)

# Calculate the magnitude
magnitude = np.sqrt(gx**2 + gy**2)

# Plot the function using a contour plot
plt.figure(figsize=(8, 6))
contours = plt.contour(x, y, z, 20, cmap='hot')
plt.clabel(contours, inline=True, fontsize=8)

# Plot the gradient vectors using quiver plot
plt.quiver(x, y, gx, gy, color='blue')

# Add details
plt.title('Gradient Field of z=x^2+y^2')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)
plt.axis('equal')  # Equal scaling for both axes
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

# Define the gradient of the function
def grad_f(x, y):
    r = np.sqrt(x**2 + y**2)
    return (x/r)*np.cos(r), (y/r)*np.cos(r)

# Create a grid of values (x, y)
N = 20
x = np.linspace(-5, 5, N)
y = np.linspace(-5, 5, N)
x, y = np.meshgrid(x, y)

# Calculate the gradient values
gx, gy = grad_f(x, y)

# Normalize the gradient vectors
magnitude = np.sqrt(gx**2 + gy**2)
magnitude[magnitude == 0] = 1  # Prevent division by zero
gx /= magnitude
gy /= magnitude

# Create subplots
fig = plt.figure(figsize=(14, 7))

# 2D gradient plot
ax1 = fig.add_subplot(121)
ax1.contour(x, y, f(x, y), 20, cmap='viridis')
ax1.quiver(x, y, gx, gy, color='blue')
ax1.set_title('2D Gradient Field')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# 3D surface and gradient plot
ax2 = fig.add_subplot(122, projection='3d')
X, Y, Z = x, y, f(x, y)
ax2.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none', alpha=0.8)
ax2.quiver(X, Y, Z, gx, gy, np.zeros_like(Z), length=0.3, normalize=True, color='red')
ax2.set_title('3D Surface and Gradient')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('f(x, y)')

# Show plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np

import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.figure_factory as ff

# Define the function and its gradient
def f(x, y):
    """Function for which we want to plot the contours and gradient."""
    return np.sin(np.sqrt(x**2 + y**2))

def grad_f(x, y):
    """Gradient of the function f."""
    r = np.sqrt(x**2 + y**2)
    with np.errstate(divide='ignore', invalid='ignore'):
        gx = x/r * np.cos(r)
        gy = y/r * np.cos(r)
        # Avoid division by zero for the center point
        gx[r == 0] = 0
        gy[r == 0] = 0
    return gx, gy

# Generate a grid of x, y points
x, y = np.meshgrid(np.linspace(-5, 5, 20), np.linspace(-5, 5, 20))

# Compute the gradient vectors u, v at each (x,y) point
u, v = grad_f(x, y)

# Calculate the function values at each point on the grid
z = f(x, y)

# Normalize the gradient vectors (for visual clarity in the quiver plot)
magnitude = np.sqrt(u**2 + v**2)
magnitude[magnitude == 0] = 1  # Prevent division by zero
u /= magnitude
v /= magnitude

# Create the contour plot
contour = go.Contour(
    x=np.linspace(-5, 5, 20),
    y=np.linspace(-5, 5, 20),
    z=z,
    colorscale='Viridis',
    name='contour',
    showscale=False
)

# Create the quiver plot for the gradient field
quiver = ff.create_quiver(
    x, y, u, v,
    scale=0.3,
    arrow_scale=0.3,
    name='quiver',
    line_width=1,
)

# Set up the figure with subplots
fig = make_subplots(
    rows=1, cols=2,
    specs=[[{'type': 'xy'}, {'type': 'scene'}]],
    subplot_titles=('2D Gradient Field with Contour', '3D Surface'),
    horizontal_spacing=0.0,
)

# Add contour to the first subplot
fig.add_trace(contour, row=1, col=1)

# Add the gradient quiver plot to the first subplot
for trace in quiver.data:
    fig.add_trace(trace, row=1, col=1)

# Add the 3D surface to the second subplot
fig_surface = go.Surface(
    x=x, y=y, z=z,
    colorscale='Viridis',
    name='surface',
)
fig.add_trace(fig_surface, row=1, col=2)

# Update the layout for better viewing
fig.update_layout(
    height=600,
    showlegend=False,
    title_text="2D Gradient Field and 3D Surface Visualization"
)




# Show the figure
fig.show()
```
